```python
import os
from topoflow.utils import rti_files
from topoflow.utils import regrid

# in and out files
input_file = '/Users/greyevenson/Projects/Bolotin/basins/WolverineGlacier/__met/TopoFlow_AORC_Alaska_Forcings.nc'

# get DEM grid info
grid_info = rti_files.read_info( '/Users/greyevenson/Projects/Bolotin/basins/WolverineGlacier/__topo/WolverineGlacier.rti', SILENT=True )

# forcings var names
var_names = ['Lwnet_tavg','Psurf_f_inst','Qair_f_inst','Rainf_tavg','Swnet_tavg','Tair_f_inst','Wind_f_inst']

# loop over vars
for var_name in var_names:

    # name the output file
    output_file = os.path.join(os.path.dirname(input_file),'TopoFlow_AORC_Alaska_Forcings_'+var_name+'.nc')
    
    # create regridded forcings file
    regrid.create_ncgs_forcings_file(var_name=var_name, nc_file_in=input_file, ncgs_file_out=output_file, grid_info=grid_info)
```
